**<font color='blue'><font size='4'>Question 1.	Load gateway_questions.csv from gateway_questions.zip. This contains data for orders sent to various gateways (connections to the exchange), and whether or not the orders succeeded (hit the exchange before competitors). Does the data suggest that any particular gateway or gateways are faster than others? (Please share your analysis code/notebook with us).**

```python
import pandas as pd
import datetime as dt
```

```python
RawData = pd.read_csv('gateway_question.csv')
RawData
```

**There are 2 approaches to find out which gateway with fast speed:**
* Approach 1 based on all cases (both successful orders and fail orders)
* Approach 2 based on cases with successful orders

# Approach 1: Both successful orders and failed orders

```python
Data_1= RawData[['gateway_targeted','timestamp']].groupby(['gateway_targeted']).agg(['count','mean'])
Data_1
```

```python
Data_1[Data_1['timestamp']['mean']==Data_1['timestamp']['mean'].min()]
```

__<font color='green'>Conclusion: In full base, gateway 1 is faster than others__

# Approach 2: Only successful orders base

```python
RawData_temp= RawData.loc[(RawData["succeeded"]==True)]
Data_2= RawData_temp[['gateway_targeted','timestamp']].groupby(['gateway_targeted']).agg(['count','mean'])
Data_2
```

```python
Data_2[Data_2['timestamp']['mean']==Data_2['timestamp']['mean'].min()]
```

__<font color='green'> Conclusion: In base including successful cases only, gateway 4 is faster than others (speed per 1 order basis)__
